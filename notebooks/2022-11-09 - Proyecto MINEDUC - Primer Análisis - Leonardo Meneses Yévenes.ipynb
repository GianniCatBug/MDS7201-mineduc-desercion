```python
print("Un " + str(no_desertores) + "% no han desertado sus carreras")
print("Un " + str(congelaron) + "% han congelado sus carreras")
print("Un " + str(abandonaron) + "% han abandonado sus carreras")
```

```python
anhos = np.sort(df_ES_UC_IP_V4["cat_periodo"].unique())
df = pd.DataFrame()
df["año"] = None
df["% no desertores"] = None
df["% congelaron"] = None
df["% abandono"] = None
for anho in anhos:
    dataframe = df_ES_UC_IP_V4[df_ES_UC_IP_V4["cat_periodo"]==anho]
    no_desertores = round((len(dataframe[dataframe["Desertor"]==0])/len(dataframe))*100,2)
    congelaron = round((len(dataframe[dataframe["Desertor"]==1])/len(dataframe))*100,2)
    abandonaron = round((len(dataframe[dataframe["Desertor"]==2])/len(dataframe))*100,2)
    new_entry = {"año": anho, '% no desertores': no_desertores , '% congelaron': congelaron, '% abandono': abandonaron}
    df = df.append(new_entry, ignore_index=True)
df
```
