## % deserción

```python
df["anio_desercion"] = df["cat_periodo"] + 1
df_desercion = df[["anio_desercion", "codigo_unico", "mrun", "desertor_1"]].drop_duplicates()
df.shape, df_desercion.shape
```

```python
n_desertores = df_desercion.groupby(["anio_desercion"]).agg(n_desertores = ("desertor_1", "sum")).reset_index()
n_matriculados = df_desercion.groupby(["anio_desercion"]).size().reset_index(name="n_matriculados")
porc_dec = pd.merge(n_desertores, n_matriculados, how="left")
porc_dec["%"] = porc_dec["n_desertores"] / porc_dec["n_matriculados"] * 100
porc_dec
```

```python
plt.figure(figsize=(12, 6))
plt.plot(porc_dec["anio_desercion"], porc_dec["%"], "-o", ms=15, color="royalblue", label="Años considerados");
plt.plot([2021, 2022], porc_dec[-2:]["%"], "-o", ms=15, color="dimgray", label="Años omitidos");
plt.title("% Deserción por año en CFT/IP");
plt.ylabel("% Deserción");
plt.xlabel("Año");

for a, porc in zip(porc_dec["anio_desercion"], porc_dec["%"]):
    correccion = -0.9 if a in [2021, 2022] else 0.5
    plt.annotate(round(porc, 1), (a, porc + correccion))

plt.legend();
```
